```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2))
my_plotter(ax1, data1, data2, {'marker':'x'})
my_plotter(ax2, data3, data4, {'marker':'o'})
```

```python
import matplotlib.pyplot as plt

car_yard = {
    'bmw': 30,
    'isuzu': 20,
    'toyota': 50,
    'Honda': 45,
    'Volks wagen': 75
}

# Sort the dictionary by values (descending)
sorted_cars = dict(sorted(car_yard.items(), key=lambda x: x[1], reverse=False))

fig, ax = plt.subplots()
ax.barh(list(sorted_cars.keys()), list(sorted_cars.values()))

ax.set_title('Car Yard Prices')
ax.set_xlabel('Price of car')
ax.set_ylabel('Type of car')

plt.show()
```

```python
shoes = {
    'Crocks':300,
    'Slippers':150,
    'Samba':2000,
    'Air Force':1500
}

h = np.random.random(1000)

n = np.linspace(0, 10, 100)

fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(12,10))

ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]

ax1.hist(h,bins=30, color='green',edgecolor='black')
ax1.set_title('Histogram')

ax2.plot(n, np.exp(n),color='red')
ax2.set_title('Line Plot')
ax2.grid(True)

ax3.bar(shoes.keys(),shoes.values(),color='orange')
ax3.set_title('Bar Chart')
ax3.grid(True)

ax4.scatter([1,3,5,7],[3,2,6,4],color='black')
ax4.set_title('Scaatter Plot')
ax4.grid(True)

plt.tight_layout()
```

```python
import seaborn as sns

df = sns.load_dataset("iris")

fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(12,10))

ax1 = axes[0, 0]
ax2 = axes[0, 1]
ax3 = axes[1, 0]
ax4 = axes[1, 1]

ax1.hist(df['sepal_length'], bins=20,color='skyblue',edgecolor='black')
ax1.set(title='Histogram of sepal length',xlabel='sepal length',ylabel='Frequency')
ax1.grid(True)

mean_by_species = df.groupby('species')['petal_length'].mean()
ax2.plot(mean_by_species.index,mean_by_species.values,marker='o',color='green')
ax2.set(title='Average petal Length by Species',xlabel='Species',ylabel='Petal Length')
ax2.grid=True

species_counts = df['species'].value_counts()
ax3.bar(species_counts.index, species_counts.values, color='orange')
ax3.set(title='Number of flowers per species',xlabel='Species',ylabel='count')
ax3.grid=True

colors = df['species'].astype('category').cat.codes
ax4.scatter(df['sepal_length'],df['petal_length'],c=colors,cmap='viridis')
ax4.set(title='Sepal Length vs Petal Length',xlabel='Sepal Length',ylabel='Petal Length')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

# Histogram with default bins
plt.hist(df['total_bill'])
plt.title("Histogram with Default Bins")
plt.show()

# Histogram with fewer bins
plt.hist(df['total_bill'], bins=5, color="orange", edgecolor="black")
plt.title("Histogram with 5 Bins")
plt.show()

# Histogram with more bins
plt.hist(df['total_bill'], bins=30, color="green", edgecolor="black")
plt.title("Histogram with 30 Bins")
plt.show()

```
